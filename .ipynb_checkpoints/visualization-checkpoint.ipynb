```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[10,20,5,25]
plt.plot(x,y)
plt.show()    
```

```python
days=["Sun","Mon","Tue","Wed","Thu","Fri","Sat"]
sales=[200,300,250,350,400,275,390]
plt.plot(days,sales)
plt.show()
```

```python
days=["Sun","Mon","Tue","Wed","Thu","Fri","Sat"]
sales=[200,300,250,350,400,275,390]
plt.plot(days,sales,marker="o",linestyle="--",color="green",markerfacecolor="black",markersize="9")
plt.title("Sales over the week")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()  
```

```python
subjects=["Maths","Science","Social","Nepali","English"]
marks=[90,84,80,75,81]
plt.bar(subjects,marks)
plt.title("Marks according to subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()  
```

```python
subjects=["Maths","Science","Social","Nepali","English"]
marks=[90,84,80,75,81]
plt.barh(subjects,marks,color="yellow")    # barh for horizontal bar
plt.title("Marks according to subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show() 
```

```python
df=pd.read_csv("sales_data.csv")
df
```

```python
ram_data=df[df['Name']=="Ram"]
plt.plot(ram_data['Month'],ram_data['Sales'])
plt.title("Ram's Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
df['Sales']=df['Sales'].replace('400s','400')
df['Sales']=pd.to_numeric(df['Sales'])               # converting into numeric form
dept_sales=df.groupby("Dept")["Sales"].sum()
print(dept_sales)
plt.bar(dept_sales.index,dept_sales.values)
plt.title("Total sales by Department")
plt.xlabel("Departments")
plt.ylabel("Sales")
plt.show()
```

```python
for name in df['Name'].unique():
    data=df[df['Name'] == name]
    plt.plot(data['Month'],data['Sales'],label=name)

plt.title("Sales comparison")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
feb=df[df['Month']=='Feb']
plt.bar(feb['Name'],feb['Sales'])
plt.title('February Sales')
plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,5,25]

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Line chart1")

b=[10,20,30,40]
plt.subplot(2,2,2)
plt.plot(x,b)
plt.title("Line chart2")

b=[10,20,30,40]
plt.subplot(2,2,3)
plt.plot(x,b)
plt.title("Line chart2")

b=[10,20,30,40]
plt.subplot(2,2,4)
plt.plot(x,b)
plt.title("Line chart2")

plt.suptitle("Dashboard")
plt.show()
```

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
ram_data=df[df['Name']=="Ram"]
plt.plot(ram_data['Month'],ram_data['Sales'])
plt.title("Ram's Sales Trend")


plt.subplot(2,2,2)
dept_sales=df.groupby("Dept")["Sales"].sum()
plt.bar(dept_sales.index,dept_sales.values)
plt.title('Dept Sales')

plt.subplot(2,2,3)
for name in df['Name'].unique():
    data=df[df['Name'] == name]
    plt.plot(data['Month'],data['Sales'],label=name)
plt.title('Dept Sales')
plt.legend()


plt.subplot(2,2,4)
plt.hist(df['Sales'])
plt.title("Sales Distribution")

plt.show()
```

```python
df=pd.DataFrame({

    "Name":["Ram","Shyam","Hari","Sita","Gita","Amit"],
    "Dept":["IT","HR","Finance","Finance","HR","IT"],
    "Sales":[200,150,300,350,180,400],
    "Age":[25,30,25,32,26,29]
})
df
```

```python
plt.scatter(df["Age"],df["Sales"])
for i in range(len(df)):
    plt.text(df["Age"][i],df["Sales"][i],df["Name"][i])
plt.title("Age VS Sales")
plt.xlabel("Age")
plt.ylabel("Sales")
plt.show()
```

```python
plt.hist(df["Sales"])
plt.show()
```

#Pie-chart

```python
dept_sales=df.groupby("Dept")['Sales'].sum()
plt.pie(dept_sales,labels=dept_sales.index,autopct="%1.1f%%")   # autopct : show data in percent
plt.show()
```

```python
# explode : spliting each dept 
explode=[0.1,0,0]
plt.pie(dept_sales,labels=dept_sales.index,autopct="%1.1f%%",explode=explode)
plt.show()
```

```python

colors=["gold","lightblue","purple"]
plt.pie(dept_sales,labels=dept_sales.index,autopct="%1.1f%%",explode=explode,colors=colors)
plt.show()
```

```python
colors=["gold","lightblue","purple"]
plt.pie(dept_sales,labels=dept_sales.index,autopct="%1.1f%%",colors=colors,startangle=90,shadow=True)
plt.show()
```

```python
plt.bar(dept_sales.index,dept_sales.values)
plt.title("Dept Performance")
plt.show()
```

```python
emp=df.groupby("Name")['Sales'].sum().sort_values()

plt.bar(emp.index,emp.values)
plt.title("Employee Performance Ranking")
plt.xticks(rotation=45)
plt.show()
```

```python
df=pd.DataFrame({
    "Name":["Pawan","Ram","Santosh","Naresh","Gita","Pooja","Sita"],
    "Dept":["IT","HR","Finance","HR","Finance","IT","Finance"],
    "Month":["Jan","Feb","March","Feb","March","Jan","Feb"],
    "Sales":[400,500,300,500,300,400,300]
})
df
```
