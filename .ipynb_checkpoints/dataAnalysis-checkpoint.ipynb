```python
import pandas as pd
import numpy as np

df=pd.read_csv("ecommerce_sales.csv")
print(df)
```

```python
# fill missing values 

df['age']=df['age'].fillna(df['age'].median())
df['rating']=df['rating'].fillna(df['rating'].mean())
df['city']=df['city'].fillna("Unknown")
df['discount']=df['discount'].fillna(0)
```

```python
df['revenue']=(df['quantity']*df['price']-df['discount'])   # add new column and calculate revenue
```

```python
df
```

```python
# to check which product generate high revenue

product_revenue=df.groupby('product')['revenue'].sum()
print("Top Product")
print(product_revenue.idxmax())  
print(product_revenue.max())
```

```python
# which city has most orders

city_order=df['city'].value_counts()
print(city_order.idxmax())  
 
```

```python
# chhecking average rating

print(f"Average Rating: {df['rating'].mean()}")
```

```python
# which category sells more (by revenue)

category_revenue=df.groupby("category")['revenue'].sum()
print(category_revenue.idxmax())  
print(category_revenue.max())
```

```python
# to show top 3-customer

customer_name=df.groupby("customer_name")['revenue'].sum()
top_customer=customer_name.sort_values(ascending=False).head(3)
top_customer
```
