```python
df=pd.read_csv("salary.csv")
print(df)
```

```python
# to display specific column 
print(df[["Name","Salary"]])
```

```python
# if NaN occur or exits we can drop or delete that rows from data or table
clean_df=df.dropna()
print(clean_df)
```

```python
# removing only specific NaN like if salary is NaN then removed that row only
print(df.dropna(subset=["Salary"]))
```

```python
# for removing NaN vaule based on column
print(df.dropna(axis=1))     # axis =1 : means for column and 0 for rows
```

```python
# feteching data using index
print(df["Name"][1])
```

```python
# loc : it returns data based on label

# syntax:     df.loc[row_label,column_label]

print(df.loc[0])
```

```python
# display only specific data like name only
print(df.loc[0,"Name"])
```

```python
# displaying like data from 1-3 index
print(df.loc[1:3])    # 3 is inclusive 
```

```python
print(df.loc[:,"Name"])     # single column
```

```python
# multiple column
print(df.loc[:,["Name","Age"]])      # ' : ' means display all rows
```

```python
# iloc : return data based on index

# suntax : df.iloc[row_index,column_index]

print(df.iloc[0])
```

```python
print(df.iloc[0,1])
```

```python
# displaying like data from 1-3 index
print(df.iloc[1:3])     # 3 is exclusive
```

```python
print(df.iloc[:,2])   # : means all rows and 2 means index 2 from column 
```

```python
# Difference between loc and iloc

  # loc : uses labels , end index inclusive 
  # iloc : uses possitions , end index exclusive
```

```python
print(df.loc[df["Salary"]>45000])   # checking single condition 
```

```python
print(df.loc[(df["Salary"]>45000) & (df["Age"]<30)])     # multiple conditions
```

```python
# increasing salary

df.loc[df["Department"]=="IT","Salary"]+=5000
print(df)
```

```python
# update values

df.loc[df["Age"]>25,"Department"]="Senoir"
print(df)
```

```python
# Adding column using loc[]

df.loc[:,"Bonus"]=df["Salary"]*0.1
print(df)
```

```python
# select all col except age

df = df.drop(columns=['Age'])   # Method 1: Using drop()
print(df)

df = df[[col for col in df.columns if col != 'Age']]    # Method 2: Using column selection
print(df)

df = df.loc[:, df.columns != 'Age']   # Method 3: Using loc
print(df)
```
