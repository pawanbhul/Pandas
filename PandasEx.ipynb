```python
df=pd.DataFrame({
    "Name":["Pawan","Santosh","Naresh","Dammer","Nirmal","Pradeep"],
    "Dept":["IT","HR","IT","Finance","HR","IT"],
    "Salary":["50000","40000","60000","45000","70000","45000"],
    "Age":["23","24","25","32","29","26"]
})
print(df)
```

```python
# groupby : split on the basis of column

df.groupby("Dept")
```

```python
for dept,data in df.groupby("Dept"):   # for loop in groupby method 
    print(dept)
    print(data)
```

```python
print(df.dtypes)
```

```python
df["Salary"] = df["Salary"].astype(int)   # changing into int 
df["Age"] = df["Age"].astype(int)         # changing into int
print(df)
print(df.dtypes)                     # checking data type
```
